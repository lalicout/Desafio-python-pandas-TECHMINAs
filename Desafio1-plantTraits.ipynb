```python
#criando uma função que retorne o número de linhas e colunas do dataset
def LinhaseColunas(dataframe):
    linhas = len(dataframe.axes[0])
    colunas= len(dataframe.axes[1])
    return(linhas,colunas)
```

```python
#utilizando a função com o dataframe
LinhaseColunas(plant)
```
